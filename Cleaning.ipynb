```python
# Dependencies
import pandas as pd
import math
```

```python
# Name of the CSV file
usa = 'Unclean/USvideos.csv'
canada = 'Unclean/CAvideos.csv'
france = 'Unclean/FRvideos.csv'
britain = 'Unclean/GBvideos.csv'
# Name the cleaned output file
outputUS = "US.csv"
outputCA = "CA.csv"
outputFR = "FR.csv"
outputGB = "GB.csv"
```

```python
usdf = pd.read_csv(usa)
frdf = pd.read_csv(france)
cadf = pd.read_csv(canada)
gbdf = pd.read_csv(britain)

#Check how many rows we have, if two are similar we assume the rest are the same
print(usdf.count())
print(gbdf.count())
usdf
```

```python
#Drop rows that have video removed
usdf = usdf[usdf['video_error_or_removed']==False]
frdf = frdf[frdf['video_error_or_removed']==False]
cadf = cadf[cadf['video_error_or_removed']==False]
gbdf = gbdf[gbdf['video_error_or_removed']==False]

#Drop columns we don't need
usdf = usdf.drop(columns = ['thumbnail_link', 'comments_disabled', 'ratings_disabled', 'video_error_or_removed']) 
frdf = frdf.drop(columns = ['thumbnail_link', 'comments_disabled', 'ratings_disabled', 'video_error_or_removed'])
cadf = cadf.drop(columns = ['thumbnail_link', 'comments_disabled', 'ratings_disabled', 'video_error_or_removed'])
gbdf = gbdf.drop(columns = ['thumbnail_link', 'comments_disabled', 'ratings_disabled', 'video_error_or_removed'])
usdf.head()
```

```python
#Tag their country
usdf['Country'] = 'US'
frdf['Country'] = 'FR'
cadf['Country'] = 'CA'
gbdf['Country'] = 'GB'

usdf.head()
```

```python
datadf = pd.concat([usdf,frdf,cadf,gbdf], axis=0)
datadf = datadf.rename(columns={'category_id':'Category ID', 'channel_title':'Channel Title','comment_count':'Comment Count','video_id':'Video ID','trending_date':'Date Trending','title':'Title','channel_title':'Channel Title','publish_time':'Time Published','views':'Num Views','likes':'Num Likes','dislikes':'Num Dislikes','tags':'Tags','description':'Description'})
```

```python
#Convert category numbers to category names
category_dict = {1:'Film & Animation',
2:'Autos & Vehicles',
10:'Music', 
15:'Pets & Animals',
17:'Sports',
18:'Short Movies',
19:'Travel & Events',
20:'Gaming',
21:'Videoblogging',
22:'People & Blogs',
23:'Comedy',
24:'Entertainment',
25:'News & Politics',
26:'Howto & Style',
27:'Education',
28:'Science & Technology',
29:'Nonprofits & Activism',
30:'Movies',
31:'Anime/Animation',
32:'Action/Adventure',
33:'Classics',
34:'Comedy',
35:'Documentary',
36:'Drama',
37:'Family',
38:'Foreign',
39:'Horror',
40:'Sci-Fi/Fantasy',
41:'Thriller',
42:'Shorts',
43:'Shows',
44:'Trailers'}

categoryList = datadf['Category ID']
categoryStr = []
name = ""

for idx,each in enumerate(categoryList):
    if math.isnan(each):
        each = 0
        categoryStr.append("N/A")
        continue
    converted = int(each)
    name = category_dict[converted]
    categoryStr.append(name)

datadf["Category Name"] = categoryStr
datadf = datadf.drop(columns=['Category ID'])
datadf
```

```python
datadf.to_csv('TotalData.csv')
```
